```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
import numpy as np

# 建立 figure 與 axes
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()

# 繪出簡單的sin函數
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x)); # 加上分號可以抑制繪圖時輸出的敘述文字
```

```python
# 使用MATLAB 介面
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
plt.plot(x, np.sin(x - 0), color='blue') # 使用顏色名字
plt.plot(x, np.sin(x - 1), color='g') # 使用顏色編碼 (rgb cmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # 0 到 1 的灰階
plt.plot(x, np.sin(x - 3), color='#FFDD44') # 16 進位編碼
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3)) # RGB 元組 值 0 到 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # 所有在HTML 中資源的顏色
```

```python
# 用樣式名稱設定
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

# 上述樣式的簡寫碼
plt.plot(x, x+4, linestyle='-') # solid
plt.plot(x, x+5, linestyle='--') # dashed
plt.plot(x, x+6, linestyle='-.') # dashdot
plt.plot(x, x+7, linestyle=':') # dotted
```

```python
# 合併線條樣式與顏色宣告 不用指定參數
plt.plot(x, x+0, '-g') # solid
plt.plot(x, x+1, '--c') # dashed
plt.plot(x, x+2, '-.k') # dashdot
plt.plot(x, x+3, ':r'); # dotted red
```

```python
# 精準調整x y 軸的顯示範圍
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
# 反向顯示 調換參數順序即可
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

```python
# 設定軸顯示的方式 使用 axis(字串)
# 讓邊界緊靠內容
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
# 確保x, y 軸比例相同
plt.plot(x, np.sin(x))
plt.axis('equal');
```

```python
# 設定標題 標籤
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
# 使用legend() 顯示兩條線的標籤
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

```python
# 使用ax.set() 一次設定多個屬性
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0,10), ylim=(-2,2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```
