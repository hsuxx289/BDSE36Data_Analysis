### 2.1 選取多個DataFrame 欄位

```python
# 取出多個欄位
movies = pd.read_csv('../../data/movie.csv')
movie_actor_director = movies[['actor_1_name','actor_2_name','actor_3_name',
                               'director_name']]
movie_actor_director.head()
```

```python
# 選取單一欄位但回傳DataFrame
print(type(movies[['director_name']]))
print(type(movies['director_name']))
```

```python
# 使用loc屬性選取多個欄位
print(type(movies.loc[:, ['director_name']]))
print(type(movies.loc[:, 'director_name']))
```

```python
# 長的串列 先指派變數
cols = ['actor_1_name','actor_2_name','actor_3_name','director_name']
movie_actor_director = movies[cols]
movie_actor_director
```

```python
# 忘了用串列 指定多欄位
movies['actor_1_name','actor_2_name','actor_3_name','director_name']
```

### 2.2 用方法選取欄位

```python
# 查看dtypes
movies.dtypes.value_counts()
```

```python
# 選取特定dtype
movies.select_dtypes(include='object')
```

```python
# number 參數 可以一次回傳 int 與float
def shorten(col):
    return (col.replace('facebook_likes', 'fb')
                .replace('_for_reviews', ''))
movies = movies.rename(columns=shorten)
movies.select_dtypes(include='number').head()
```

```python
# 用字串選取兩種型別
movies.select_dtypes(include=['int', 'object']).head()
```

```python
# exclude
movies.select_dtypes(exclude='float').head()
```

```python
# like
movies.filter(like='fb').head()
```

```python
# item
cols = ['actor_1_name', 'actor_2_name', 'actor_3_name', 'direcotr']
movies.filter(items=cols).head()
```

```python
# regex
movies.filter(regex=r'\d').head()
```
