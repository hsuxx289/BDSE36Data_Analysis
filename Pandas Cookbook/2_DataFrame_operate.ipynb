## 2.3 對欄位名稱進行排序

```python
# 取出資料集 並縮減部分欄位名稱
movies = pd.read_csv('../../data/movie.csv')
def shorten(col):
    return (col.replace('facebook_likes', 'fb')
                .replace('_for_reviews', ''))
movies = movies.rename(columns=shorten)
```

```python
# 檢視欄位名稱跟順序
movies.columns
```

```python
# 分類欄位名稱
cat_core = ['movie_title', 'title_year', 'content_rating', 'genres']
cat_people = ['director_name', 'actor_1_name', 'actor_2_name', 'actor_3_name']
cat_other = ['color', 'country', 'language', 'plot_keywords', 'movie_imdb_link']
cont_fb = ['director_fb', 'actor_1_fb', 'actor_2_fb', 'actor_3_fb', 'cast_total_fb', 'movie_fb']
cont_finance = ['budget', 'gross']
cont_num_reviews = ['num_voted_users', 'num_user', 'num_critic']
cont_other = ['imdb_score', 'duration', 'aspect_ratio', 'facenumber_in_poster']
```

```python
# 檢查有沒有遺漏欄位
new_col_order = cat_core + cat_people + cat_other + cont_fb + cont_finance + cont_num_reviews + \
                cont_other

set(movies.columns) == set(new_col_order)
```

```python
# 直接用新的欄位順序的fancy index
movies[new_col_order].head()
```

## 2.6 DataFrame 的算符運算

```python
# 載入大學資料集
colleges = pd.read_csv('../../data/college.csv')

# 資料型別不是數值 會報錯
colleges + 5
```

```python
# 挑選適合的欄位
colleges = pd.read_csv('../../data/college.csv', index_col='INSTNM')
college_ugds = colleges.filter(like='UGDS_')
college_ugds.head()
```

```python
# 隨機挑一間大學
name = 'Northwest-Shoals Community College'
college_ugds.loc[name]
```

```python
# 第二項 0.125 用round()沒有進位
college_ugds.loc[name].round(2)
```

```python
# 加上0.0001 向上進位
(college_ugds.loc[name] + .0001).round(2)
```

```python
# 為了floor 不要出錯 加上 0.00501
college_ugds + 0.00501
```

```python
# 使用 // floor 向下捨去
(college_ugds + 0.00501) // 0.01
```

```python
# 再除以100 恢復原來的單位
college_ugds_op_round = (college_ugds + 0.00501) // 0.01 / 100
college_ugds_op_round
```

```python
# 比較兩者
college_ugds_round = (college_ugds + 0.00001).round(2)

college_ugds_round.equals(college_ugds_op_round)
```

## 2.7 比較缺失值

```python
# np.nan 比較運算都會回傳False
print(np.nan == np.nan)
print(np.nan > 5)
print(5 > np.nan)

# 除了 !=
print(np.nan != np.nan)
```

```python
# 挑選適合的欄位
colleges = pd.read_csv('../../data/college.csv', index_col='INSTNM')
college_ugds = colleges.filter(like='UGDS_')
college_ugds.head()
```

```python
# == 用於純量比較
college_ugds == 0.0019
```

```python
# 自己跟自己比
college_self_compare = college_ugds == college_ugds
college_self_compare.head()
```

```python
# 用all() 檢查缺失值
college_self_compare.all()
```

```python
# 但如果用 == 檢查 就會出錯 因為不能比較
(college_ugds == np.nan).sum()
```

```python
# 改用 .isna()
college_ugds.isna().sum()
```

```python
# 使用.equals()
college_ugds.equals(college_ugds)
```

```python
# .eq() 跟 .equals() 不同
college_ugds.eq(0.0019).head()
```

```python
# 單元測試
from pandas.testing import assert_frame_equal
assert_frame_equal(college_ugds, college_ugds) is None
```

## 2.9 案例演練: 確定大學校園的多樣性

```python
# 查看美國新聞統計的多樣性前10名
print(pd.read_csv('../../data/college_diversity.csv', index_col='School'))
```

```python
# 挑選適合的欄位
colleges = pd.read_csv('../../data/college.csv', index_col='INSTNM')
college_ugds = colleges.filter(like='UGDS_')
```

```python
# 查看缺失值並排序
(college_ugds.isna()
             .sum(axis='columns')
             .sort_values(ascending=False)
             .head()
)
```

```python
# 全部都缺就移除
college_ugds = college_ugds.dropna(how='all')
college_ugds.isna().sum()
```

```python
# 找出大於15% 的種族
college_ugds.ge(0.15)
```

```python
# 計算每個學校 大於15%的種族個數
diversity_metric = college_ugds.ge(0.15).sum(axis='columns')

diversity_metric.head()
```

```python
# 檢視分布狀況
diversity_metric.value_counts()
```

```python
# 找出特別高的兩間
diversity_metric.sort_values(ascending=False).head()
```

```python
# 查看原始資訊
print(college_ugds.loc[['Central Texas Beauty College-Temple',
                  'Regency Beauty Institute-Austin']])
```

```python
# 比較美國新聞前5名
us_news_top = ['Rutgers University-Newark',
               'Andrews University',
               'Stanford University',
               'University of Houston',
               'University of Nevada-Las Vegas']

print(diversity_metric.loc[us_news_top])
```

```python
# 找最大種族占比
(college_ugds
    .max(axis=1)
    .sort_values(ascending=False)
    .head(10)
)
```

```python
# 找是否有學校所有種族都超過1%
college_ugds[(college_ugds > 0.01).all(axis=1)]
```
